# Congrats! You just retreived an image from the Google Streetview API.

##  Let's deconstruct the URL

    1) Root URL:             https://maps.googleapis.com/maps/api/streetview

    2) Question Mark:        ?

    3) Image Size:           size=800x800

    4) Ampersand:            &

    5) Location:             location=40.8078647,-73.9603171

    6) Ampersand:            &

    7) Heading:              heading=200

    8) Ampersand:            &

    9) Pitch:                pitch=0
    
## What are these parameters doing?

Well,

    1) The root URL establishes where on the internet the API lives.

    2) The question mark indicates to the API that one or several parameters are coming.

    3) Image size is our first paremter. It indicates how big we want the image to be (maximum size is 800 x 800)

    4) "&" is used to chain together multiple parameters into a single API request

    5) Location indicates from where we want the image (Latitude, Longitude)

    6) Another "&" to chain together the next parameter

    7) Heading indicates which direction (from left to right) we want the camera to be facing

    8) Another "&" to chain together the next parameter

    9) Pitch indicates which directino (from down to up) we want the camera to be facing
    
Pretty cool, right?

# Make a function to retrieve an image based on lat, lon, heading and pitch

```python
def get_image(lat, lon, heading, pitch):
    
    # Copy and paste the URL below, replacing lat, lon, heading and pitch with '{}'
    template = "https://maps.googleapis.com/maps/api/streetview?size=800x800&location={},{}&heading={}&pitch={}"
    
    # Insert lat, lon, heading and pitch into the url using string formatting
    url = template.format(lat, lon, heading, pitch)
    
    # Make the API request
    response = requests.get(url)
    
    # Convert it to an image
    img = Image.open(requests.get(url, stream=True).raw)
    
    # Display the image
    return img
```

```python
get_image(40.7735772,-73.9705028, 270, 0)
```

# Plan of attack: Write a for loop to cycle through each lat, lon coordinate and download the corresponding image from streetview

# My solution:

```python
import os
```

```python
def get_images(geojson, heading, pitch):
    
    # Keep track of image count for file naming
    count = 0
    
    # For every coordinate in the geojson
    for i in geojson['features'][0]['geometry']['coordinates']:

        # Get the latitude
        lat = i[1]
        
        # Get the longitude
        lon = i[0]
        
        # Use our get_image function to get the image
        img = get_image(lat, lon, heading, pitch)
        
        # Define the image filename as 0001, 0002, 0003, etc
        img_filename = str(count).zfill(4) + ".jpg"
        
        # Create an output folder to save our images
        if not os.path.exists("output"):
            os.makedirs("output")
        
        # Get our current working directory
        current_dir = os.getcwd()
        
        # Define output directory
        output_dir = os.path.join(current_dir,"output")
        
        # Define the filepath
        filepath = os.path.join(output_dir, img_filename)
        
        # Save the image to output directory
        img.save(open(filepath, 'w'))
        
        # Increase the count by 1
        count += 1
```

```python
get_images(geojson, 90, 0)
```

# Stitch these images together to make a GIF

You can upload the images to https://ezgif.com/maker (easier).

Or you can do it in Python. We will use a library called [MoviePy](https://zulko.github.io/moviepy/index.html).

```python
# Install moviepy
!conda install -c conda-forge moviepy --yes
```

```python
import glob
import moviepy.editor as mpy
```

Define the gif name and frames per second (fps).

```python
gif_name = "my_gif"
fps = 5
```

Get a list of all the jpg files.

```python
file_list = glob.glob('output/*.jpg')
```

Combine the jpg files in an image sequence clip using MoviePy.

```python
clip = mpy.ImageSequenceClip(file_list, fps=fps)
```

Write the image sequence clip to a gif.

```python
clip.write_gif('{}.gif'.format(gif_name), fps=fps)
```

View the gif.

```python
from IPython.display import HTML
HTML('<img src="{}.gif">'.format(gif_name))
```
